```python
import numpy
import matplotlib.pyplot as plt 
```

```python
def plot_chart(x, y):
    plt.xlabel("Elements")
    plt.ylabel("Equivalent Stress [MPa]")
    plt.grid()
    plt.ticklabel_format(axis="x", style="sci", scilimits=(0,0))

    plt.plot(x, y, 'o-', color='red')
    plt.show()
```

```python
x = [416561, 456018,547220, 755277]
y = [24.587, 26.731, 27.869, 28.243]

plot_chart(x,y)
plt.savefig('convergence.png', format='png')
```
